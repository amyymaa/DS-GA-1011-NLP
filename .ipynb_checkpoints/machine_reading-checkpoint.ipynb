## 1. Random QA

1) What is the probability $p_{k,n}$ that our random QA system will output an answer with $k$ tokens when given a context paragraph with $n$ token as input?

$$p_{k,n} = \frac{n+1-k}{\Sigma_{i=1}^n (n+1-i)} = \frac{2(n+1-k)}{(n+1)n} $$

There are n+1-k kinds of k tokens which can be generated from n tokens.

And there are n+1-i kinds of possibility for s = i, i is a integer from 1 to n.

So there are $\Sigma_{i=1}^n (n+1-i)$ kinds of possible token sequences.

Therefore, the probability of output an answer with k tokens when given n tokens is $p_{k,n} = \frac{n+1-k}{\Sigma_{i=1}^n (n+1-i)}$

2) For a fixed $n$ of $100$, produce a plot of $p_{k,n}$ and $\hat{p}_{k,n}$  vs $k$. Design a monte-carlo experiment to estimate the values for $\hat{p}_{k,n}$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def P_k_n(k, n):
    #implement answer from part 1 here
    #print([n+1-i for i in range(1,n+1)])
    p_k_n = (n+1-k)/sum([n+1-i for i in range(1,n+1)])
    return p_k_n
```

```python
def monte_carlo_p_k_n(k, n, T = int(1e5)):
    #T is the number of times you run experiment
    num_k = 0
    pairs = []
    for s in range(1, n+1):
      for e in range(s, n+1):
        pairs.append([s,e])
    for t in range(T):
      index = int(np.random.randint(0, len(pairs), 1))
      pair = pairs[index]
      s, e = pair[0], pair[1]
      if e-s == k-1:
        num_k += 1
    p_k_n_hat = num_k/T
    return p_k_n_hat
```

```python
k_array = np.arange(1, 100, 5)
p_k_n_hat = [monte_carlo_p_k_n(k, n=100, T = int(1e5)) for k in k_array]
p_k_n = [P_k_n(k, n=100) for k in k_array]
```

```python
print(p_k_n_hat)
print([round(p_k_n[i], 5) for i in range(len(k_array))])
```

```python
plt.plot(k_array, p_k_n_hat, label = r'$\hat{p_{k,n}}$')
plt.plot(k_array, p_k_n, label = r'$p_{k, n}$')
plt.legend()
```

3) Calculate the expected length of the answer given by your random QA system. i,e write an expression for $L_n = E[K]$ (It's fine to leave it as a summation).

$$ L_n = \Sigma_{k=1}^n p_{k,n} k = \Sigma_{k=1}^n\frac{(n+1-k)k}{\Sigma_{i=1}^n (n+1-i)} = \Sigma_{k=1}^n[\frac{2(n+1-k)k}{(n+1)n}]$$

4) Plot $\hat{L_n}$ estimated through monte-carlo simulations and $L_n$ for $n = 5, 10, 25, 50, 100, 250$ and $500$.

```python
def L_n(n):
    #implement answer from above
    l_n = 0
    for k in range(1, n+1):
      length = (n+1-k)*k/sum([n+1-i for i in range(1,n+1)])
      l_n += length
    return l_n

def monte_carlo_L_n(n, T = int(1e5)):
    #T is the number of times you run experiment
    l_n = 0
    pairs = []
    for s in range(1, n+1):
      for e in range(s, n+1):
        pairs.append([s,e])
    for t in range(T):
      index = int(np.random.randint(0, len(pairs), 1))
      pair = pairs[index]
      s, e = pair[0], pair[1]
      k = e-s+1
      l_n += k

    l_n = l_n/T
    return l_n
```

```python
n_array = [5, 10, 25, 50, 100, 250, 500]
l_n_hat = [monte_carlo_L_n(n, T = int(1e5)) for n in n_array]
l_n = [L_n( n) for n in n_array]
```

```python
print(l_n_hat)
print([round(l_n[i], 5) for i in range(len(n_array))])
```

```python
plt.plot(n_array, l_n_hat, '-o', label = r'$\hat{L_n}$')
plt.plot(n_array, l_n,  '-*', label = r'$L_n$')
plt.legend()
```

```python
def estimate_alpha(list_of_answers):
  starts, ends = list_of_answers[0], list_of_answers[1]
  no_answer = 0
  all_answer = len(list_of_answers[0])
  for i in range(len(dataset)):
    if starts[i] == 0 and ends[i] == 0:
      no_answer += 1
  alpha = no_answer/all_answer
  return alpha
```

```python
## implement random QA model
def random_QA(context, alpha):
  context = context[context.nonzero()]
  n = len(context)
  #print(n)
  ## calculate the distribution
  pairs = []
  for s in range(1, n+1):
    for e in range(s, n+1):
      pairs.append([s,e])
  probability = [(1-alpha)/len(pairs) for i in range(len(pairs))]
  pairs.append([0,0])
  probability.append(alpha)
  index = int(np.where(np.random.multinomial(1,probability)==1)[0])
  #print(index)
  pred_answer = pairs[index]
  return pred_answer

## find F1 and EM on Dev Set
from collections import Counter
def compute_f1(a_gold, a_pred):
    # gold_toks: list of 2
    # pred_toks: list of 2
    gold_toks = range(a_gold[0], a_gold[1]+1)
    pred_toks = range(a_pred[0], a_pred[1]+1)
    common = Counter(gold_toks) & Counter(pred_toks)
    num_same = sum(common.values())
    # if len(gold_toks) == 1 or len(pred_toks) == 1:
    #     # If either is no-answer, then F1 is 1 if they agree, 0 otherwise
    #     return int(gold_toks == pred_toks)
    if num_same == 0:
        return 0
    precision = 1.0 * num_same / len(pred_toks)
    recall = 1.0 * num_same / len(gold_toks)
    f1 = (2 * precision * recall) / (precision + recall)
    return f1

def compute_em(a_gold, a_pred):
    return int(a_gold == a_pred)
```

**Reason for Alpha Selection**

Theoretically, we cannot use a parameter estimated on dev set to predict on dev set and report scores, which is a cheat. **However, in our Random Q&A, there is no parameter learning.** In fact we can even estimate the EM score by derivation.

Suppose $\beta$ represents the real probability that a question has no answer in context, and we randomly sample with the estimated $\alpha$. Assume a sentence has N tokens, so it has $\frac{N(N+1)}{2}$ pairs of potential answer. then the probability that we can predict the correct answer is 
$$\alpha\beta+(1-\beta)\frac{2(1-\alpha)}{N(N+1)} = \frac{2(1-\alpha -\beta)+(N(N+1)+2)\alpha\beta}{N(N+1)}$$

Generally, N is among 40 to 100. If we take 1st order derivative of $\alpha$, we have $\frac{-2+(N(N+1)+2)\beta}{N(N+1)} = \beta +\frac{2(\beta-1)}{N(N+1)}$, so correct rate is linear to $\alpha$, which means we have best scores either when $\alpha = 0$ or $\alpha = 1$. As long as $\beta > \frac{2}{N(N+1)+2}$, which consists with our dataset, the probability of correct prediction is always increase in $\alpha$. We would have the best scores when $\alpha$ is 1, rather than when $\alpha = \beta$. 

Therefore, I think we can use $\alpha$ estimated from dev set.
